```python
# - Viết hàm tính các chuẩn p 1,2, inf cho 50 dòng đầu tiên của array trong mục 1.


# Ham tinh chuan p = 1 , 2, inf
def matrix_norm(array, p):
    # Kiem tra except neu array khong phai 50 dong

    if array.shape[0] != 50:
        raise Exception("Array is not 50 rows")

    # Tao mot dataframe 50 cot voi columns ten la tu 1, 2, ..., 50
    df = pd.DataFrame(columns = range(0, 50))

    # Tinh cac gia tri chuan va dua vao dataframe
    for i in range(50):
        for j in range(50):
            if i == j:
                df.loc[i, j] = 0
            elif i > j:
                df.loc[i, j] = df.loc[j, i]
            else:
                dist = np.linalg.norm(array[i, :] - array[j, :], p)
                df.loc[i, j] = dist
    return df
    
```

```python
# Chuẩn p = 1
array_1 = array[0:50, :]
df_1 = matrix_norm(array_1, 1)
```

```python
# Chuẩn p = 2
array_2 = array[0:50, :]
df_2 = matrix_norm(array_2, 2)
df_2
```

```python
# Chuẩn p = inf
array_inf = array[0:50, :]
df_inf = matrix_norm(array_inf, np.inf)
df_inf
```

#### 2.2 Do do tuong dong don gian

```python
# Ham do su tuong dong don gian
def do_do_tuong_dong(x,y):
    # sim = sum(1 for i in range(len(x)) if x[i] == y[i])
    sim = np.sum(x == y)
    return sim

def mattrix_simple_similar(array):

    # khoi tao dataframe voi 100 cot
    df = pd.DataFrame(columns = range(0, 100))

    len_columns = array.shape[1]

    for i in range(100):
        for j in range(100):
            if i == j:
                df.loc[i, j] = len_columns # Phần tử trên đường chéo là số lượng các cột vì chúng hoàn toàn giống nhau
            elif i > j:
                df.loc[i, j] = df.loc[j, i] # Tính chất đối xứng của ma trận tương đồng
            else:
                sim = do_do_tuong_dong(array[i, :], array[j, :]) # Tính độ tương đồng giữa hai dòng i và j
                df.loc[i, j] = sim
    return df
```

```python
array_100 = array[:100, :]
matrix_sim = mattrix_simple_similar(array_100)
```

```python
matrix_sim.head()
```

```python
# Ham tìm láng giềng gần nhất, áp dụng cho cả ma trận tương đồng đơn giản và ma trận tương đồng tần suất

import pandas as pd
import numpy as np

def tim_lang_gieng_gan_nhat(matrix_similar, k=5): # Mặc định k = 5 nếu không có tham số truyền vào

    """
        Tìm k láng giềng gần nhất cho mỗi dòng trong ma trận
        matrix_similar: ma trận tương đồng đơn giản hoặc ma trận tương đồng tần suất
        k: số lượng láng giềng gần nhất cần tìm (mặc định là 5)
        Trả về một dictionary với key là chỉ số dòng và value là danh sách các chỉ số dòng của k láng giềng gần nhất
    """
    neighbors = {}

    for i in range(matrix_similar.shape[0]):
        # Sắp xếp theo giá trị độ tương đồng giảm dần và lấy k dòng gần nhất (bỏ chính nó)
        sorted_neighbors = matrix_similar.loc[i].sort_values(ascending=False).index.tolist()
        sorted_neighbors.remove(i)  # Bỏ chính nó
        neighbors[i] = sorted_neighbors[:k]  # Lấy k láng giềng gần nhất
    
    return neighbors

```

```python
# Tìm 8 láng giềng gần nhất cho mỗi dòng trong ma trận tương đồng đơn giản
# matrix_sim = mattrix_simple_similar(array_100)
# matrix_sim đã được tính ở trên
k_neighbors = tim_lang_gieng_gan_nhat(matrix_sim, k=8)

# In ra 5 láng giềng gần nhất của dòng đầu tiên và các giá trị tương đồng của nó
print("5 láng giềng gần nhất của dòng thu 10 cua ma tran tuong dong don gian:")
print(k_neighbors[9])
```

```python
# Hàm độ đo tần suất xuất hiện ngược
import numpy as np
import pandas as pd

def tinh_tan_suat(array):
    """ Tính tần suất xuất hiện của từng giá trị trong mỗi cột
        array: ma trận đầu vào
        Trả về một dictionary với key là chỉ số cột và value là một dictionary với các giá trị duy nhất và tần suất tương ứng
    """
    tan_suat = {}
    for col in range(array.shape[1]):
        # unique: các giá trị duy nhất trong cột
        # counts: số lần xuất hiện của từng giá trị
        # freqs: tần suất xuất hiện của từng giá trị
        unique, counts = np.unique(array[:, col], return_counts=True)
        freqs = counts / len(array)  # Tỉ lệ xuất hiện
        tan_suat[col] = dict(zip(unique, freqs))
    return tan_suat

def do_do_tan_suat_xuat_hien_nguoc(x, y, tan_suat):
    """ Tính độ đo giữa hai dòng theo công thức đã cho """
    sim = 0
    for k in range(len(x)):
        if x[k] == y[k]:  # Nếu giống nhau, áp dụng công thức
            pk = tan_suat[k].get(x[k], 1)  # Lấy tần suất hoặc mặc định là 1 nếu không tìm thấy
            sim += 1 / (pk ** 2)
    return sim

def ma_tran_tuong_dong_tan_suat(array):
    """ Tạo ma trận tương đồng dựa trên độ đo tần suất xuất hiện ngược """
    df = pd.DataFrame(columns=range(0, 100))
    
    tan_suat = tinh_tan_suat(array)
    num_rows = min(100, array.shape[0])  # Chỉ tính cho 100 dòng đầu tiên

    for i in range(num_rows):
        for j in range(num_rows):
            if i == j:
                df.loc[i, j] = sum(1 / (tan_suat[k].get(array[i, k], 1) ** 2) for k in range(array.shape[1]))
            elif i > j:
                df.loc[i, j] = df.loc[j, i]  # Ma trận đối xứng
            else:
                sim = do_do_tan_suat_xuat_hien_nguoc(array[i, :], array[j, :], tan_suat)
                df.loc[i, j] = sim

    return df
```

```python
# Tính ma trận tương đồng tần suất cho 100 dòng đầu tiên
# array_100 = array[:100, :] # 100 dòng đầu tiên


array_100 = array[:100, :]
matrix_frequency = ma_tran_tuong_dong_tan_suat(array_100)
matrix_frequency.head()
```

```python
# Tìm 8 láng giềng gần nhất cho mỗi dòng trong ma trận tương đồng tần suất
# matrix_frequency = ma_tran_tuong_dong_tan_suat(array_100)
# matrix_frequency đã được tính ở trên

k_neighbors = tim_lang_gieng_gan_nhat(matrix_frequency, k=8)

# In ra 5 láng giềng gần nhất của dòng đầu tiên và các giá trị tương đồng của nó
print("8 láng giềng gần nhất của dòng thứ 10 trong ma trận tương đồng tần suất là:")
print(k_neighbors[9]) # Dòng thứ 10 trong Python là chỉ số 9
```
